```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
actual_values = np.array([100, 150, 200, 250])
```

```python
predicted_values = np.array([90, 140, 180, 270])
```

```python
mse=sum((predicted_values-actual_values)**2)/len(predicted_values)
```

```python
print(mse)
```

```python
mae=sum(abs(predicted_values-actual_values))/len(predicted_values)
```

```python
print(mae)
```
